```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
sns.set(color_codes=True)
%matplotlib inline
```

```python
# loading dataset
df = pd.read_csv('Tokyo Medals 2021.csv')
df.shape
```

```python
df.head()
```

```python
# top 5 countries by Rank
top5_country = df.sort_values(by='Rank By Total', ascending=True).head()
top5_country
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=top5_country, x='Total', y='Country')
```

```python
# top 5 countries by Gold Medal
top5_gold_medal = df.sort_values(by='Gold Medal', ascending=False).head().loc[:, ['Country', 'Gold Medal']]
top5_gold_medal
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=top5_gold_medal, y='Country', x='Gold Medal')
```

```python
# top 5 countries by Silver Medal
top5_silver_medal = df.sort_values(by='Silver Medal', ascending=False).head().loc[:, ['Country', 'Silver Medal']]
top5_silver_medal
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=top5_silver_medal, y='Country', x='Silver Medal')
```

```python
# top 5 countries by Bronze Medal
top5_bronze_medal = df.sort_values(by='Bronze Medal', ascending=False).head().loc[:, ['Country', 'Bronze Medal']]
top5_bronze_medal
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=top5_bronze_medal, x='Bronze Medal', y='Country')
```

```python
# getting countries Gold, Silver and Bronze medals
def get_countryMedals(country):
    country_medals = df[['Country', 'Gold Medal', 'Silver Medal', 'Bronze Medal']][df['Country']==country]
    country_medals.plot(kind='bar', color=['gold', 'silver', 'brown'])
    plt.title(f'{country} ({country_medals.values[0][1:].sum()})', fontsize=15)
    plt.xlabel('Medals')
    return country_medals
```

```python
get_countryMedals('ROC')
```

```python
# countries with zero Gold Medals
zero_GoldMedal = df['Country'][df['Gold Medal'] == 0].values
for i in zero_GoldMedal:
    print(i, end=", ")
```

```python
# countries with zero Silver Medals
zero_SilverMedal = df['Country'][df['Silver Medal'] == 0].values
for i in zero_SilverMedal:
    print(i, end=", ")
```

```python
# countries with zero Bronze Medals
zero_BronzeMedal = df['Country'][df['Bronze Medal'] == 0].values
for i in zero_BronzeMedal:
    print(i, end=", ")
```

```python
# All countries by total medal
plt.figure(figsize=(20, 20))
sns.barplot(data=df.sort_values(by='Total', ascending=False), x='Total', y='Country')
```

```python
# All countries by Gold medal
plt.figure(figsize=(20, 20))
sns.barplot(data=df[df['Gold Medal']>0].sort_values(by='Gold Medal', ascending=False),
            x='Gold Medal', y='Country', palette='RdYlBu')
```

```python
# All countries by Silver medal
plt.figure(figsize=(20, 20))
sns.barplot(data=df[df['Silver Medal']>0].sort_values(by='Silver Medal', ascending=False),
            x='Silver Medal', y='Country', palette='PuRd_r')
```

```python
# All countries by Bronze medal
plt.figure(figsize=(20, 20))
sns.barplot(data=df[df['Bronze Medal']>0].sort_values(by='Bronze Medal', ascending=False),
            x='Bronze Medal', y='Country', palette='RdYlBu')
```
